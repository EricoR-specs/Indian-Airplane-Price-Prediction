```python
import numpy as np
import pandas as pd
import pickle
import json
```

```python
with open('best_xgb_model.pkl', 'rb') as file_1:
    best_xgb = pickle.load(file_1)
with open('selected_features.txt', 'r') as file_2:
    features = json.load(file_2)
```

```python
print("features:", features)
```

```python
import pandas as pd
import numpy as np

# Daftar data untuk diacak
airlines = ['SpiceJet', 'AirAsia', 'Vistara', 'IndiGo', 'Akasa Air']
flight_numbers = [f'SG-{np.random.randint(1000, 9999)}' for _ in range(5)] + \
                 [f'I5-{np.random.randint(100, 999)}' for _ in range(5)] + \
                 [f'UK-{np.random.randint(900, 999)}' for _ in range(5)] + \
                 [f'6E-{np.random.randint(100, 999)}' for _ in range(5)] + \
                 [f'QP-{np.random.randint(100, 999)}' for _ in range(5)]
source_cities = ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Chennai']
destination_cities = ['Mumbai', 'Delhi', 'Bangalore', 'Kolkata', 'Chennai']
departure_times = ['Early_Morning', 'Morning', 'Afternoon', 'Evening', 'Night']
arrival_times = ['Early_Morning', 'Morning', 'Afternoon', 'Evening', 'Night']
stops = ['zero', 'one', 'two_or_more']
classes = ['Economy', 'Business']
durations = np.round(np.random.uniform(1.5, 4.0, 5), 2)
days_left = np.random.randint(1, 60, 5)

# Membuat DataFrame
data = {
    'airline': np.random.choice(airlines, 5),
    'flight': np.random.choice(flight_numbers, 5),
    'source_city': np.random.choice(source_cities, 5),
    'departure_time': np.random.choice(departure_times, 5),
    'stops': np.random.choice(stops, 5),
    'arrival_time': np.random.choice(arrival_times, 5),
    'destination_city': np.random.choice(destination_cities, 5),
    'class': np.random.choice(classes, 5),
    'duration': durations,
    'days_left': days_left,
}

df = pd.DataFrame(data)

print(df)
```

```python
predict_data = df[features]
predictions = best_xgb.predict(predict_data)
```

```python
print("Prediction of the model:", predictions)
```
